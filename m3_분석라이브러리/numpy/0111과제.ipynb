Q3_0111. 1차원 배열을 생성한 후 -1을 이용해서 차원을 변경해 보세요
- 1차원 -> 2,3차원
- 2차원 -> 1,3차원

```python
# -1 을 쓰면 알아서 계산 해줌

arrq3 = np.arange(18)
arrq32 = arrq3.reshape(2,-1)
arrq33 = arrq3.reshape(3,2,-1)
print(arrq3, '\n')
print(arrq32, '\n')
print(arrq33, '\n')
```

```python
arrq34 = arrq32.reshape(-1,)
arrq35 = arrq32.reshape(3,2,-1)
print(arrq34, '\n')
print(arrq35, '\n')
```

```python
arrq36 = arrq33.reshape(-1,)
arrq37 = arrq33.reshape(2,-1)
print(arrq36, '\n')
print(arrq37, '\n')
```

Q6_0111. 두 배열 [[1,2],[3,4]]와 [[5,6],[7,8]]를 가로로 연결하기

```python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
concatenated = np.concatenate((a,b), axis=1)
print(concatenated, '\n')
```

Q7_0111. 두 배열 [[1,2,3]]와 [[4,5,6]]을 세로로 연결하기

```python
a = np.array([[1,2,3]])
b = np.array([[4,5,6]])
concatenated = np.concatenate((a,b), axis=0)
stacked = np.vstack((a,b))
print(concatenated, '\n')
print(stacked)
```

```python
#Q11_0111. 1~100까지 정수에서 5의 배수이면서 2의 배수인 것만을 출력(for문과 배열 두 가지 방식)
```

```python
array_f = np.arange(1,101)
arrayq11 = array_f[(array_f % 5 == 0) & (array_f % 2 == 0)]
print(arrayq11)
```

```python
array_f = np.arange(1, 101)
arrayq11 = []

for num in array_f:
    if num % 5 == 0 and num % 2 == 0:
        arrayq11.append(num)

arrayq11 = np.array(arrayq11)
print(arrayq11)
```
