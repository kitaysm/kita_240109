```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/ktd231026/m3_분석라이브러리/시각화/주가데이터.csv')
df.head()
```

```python
# 칼럼 선택 : 'Close', 'Start', 'High', 'Low'
df['NDate'] = pd.to_datetime(df.Date)
df1 = df.set_index('NDate')
df1.drop(['Date','Volume'], axis=1, inplace=True)
df1.head()
```

# Q1_0115. df1에 대하여 Close 컬럼으로 다음을 수행하세요.
- NDate 별 막대그래프
- 6월만을 추출하여 Close 칼럼으로 일별 막대 그래프를 표현
- 히스토그램

```python
import pandas as pd
import matplotlib.pyplot as plt

# 전체 데이터에 대한 'Close' 칼럼의 막대 그래프
plt.figure(figsize=(10, 6))
df1['Close'].plot(kind='bar')
plt.title('Daily Closing Prices')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.show()

# 6월 데이터만 추출
df1_june = df1[df1.index.month == 6]

# 6월 데이터에 대한 'Close' 칼럼의 막대 그래프
plt.figure(figsize=(10, 6))
df1_june['Close'].iloc[::-1].plot(kind='bar', color='green')
plt.title('Daily Closing Prices for June')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.show()

# 히스토그램 (Close 칼럼)
plt.figure(figsize=(10, 6))
df1['Close'].hist(bins=10)
plt.title('Histogram of Closing Prices')
plt.xlabel('Close Price')
plt.ylabel('Frequency')
plt.show()
```
