Q1_0117. 타이타닉 데이터셋(sns.load_dataset("titanic"))을 사용하여, 클래스(class)별 승객 수를 나타내는 막대 그래프를 생성하세요

각 막대 위에 승객 수 표시

```python
df = sns.load_dataset('titanic')
counts = df['class'].value_counts()
plt.figure(figsize=(8,6))
barplot = sns.barplot(x=counts.index, y=counts.values)
barplot.bar_label(barplot.containers[0])
plt.title('Passenger Count by Class on Titanic')
plt.xlabel('Class')
plt.ylabel('Number of Passengers')
plt.show()
```

Q2_0117. sns.load.dataset('flights')을 로드하고, 이를 피벗 테이블로 변환하여 연도별 월간 승객 수를 나타내세요. 이 데이터를 사용하여 히트맵을 생성하세요. 색상 맵은 자유롭게 선택

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 'flights' 데이터셋 로드
flights = sns.load_dataset('flights')

# 피벗 테이블 생성
pivot_flights = flights.pivot("month", "year", "passengers")

# 히트맵 생성
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_flights, annot=True, fmt="d", cmap="YlGnBu")

# 제목 설정
plt.title('Monthly Passengers by Year (Flights Dataset)')

# 그래프 표시
plt.show()
```

Q3_0117. sns.load_dataset('penguins')을 사용하여 페어플롯(pairplot)을 생성하세요. 종(species)에 따라 다른 색상을 사용

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 'penguins' 데이터셋 로드
penguins = sns.load_dataset('penguins')

# 페어플롯 생성, 종(species)에 따라 색상 구분
sns.pairplot(penguins, hue='species', palette='rainbow')

# 제목 설정
plt.suptitle('Pairplot of Penguins Dataset by Species', y=1.02)  # y=1.02는 제목이 그래프와 겹치지 않도록 조정

# 그래프 표시
plt.show()
```

Q4_0117. 타이타닉 데이터셋을 사용하여, 객실 등급(class)에 따른 탑승객의 나이를 나타내는 박스 플롯과 바이올린 플롯을 같은 그래프 내에서 나란히 생성하세요. 이를 통해 각 클래스별 나이 분포의 차이를 비교하세요

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 타이타닉 데이터셋 로드
titanic = sns.load_dataset("titanic")

# 그래프 생성을 위한 그림판 설정
plt.figure(figsize=(12, 6))

# 박스 플롯 생성
plt.subplot(1, 2, 1)
sns.boxplot(x="class", y="age", data=titanic)
plt.title("Box Plot of Age by Class")

# 바이올린 플롯 생성
plt.subplot(1, 2, 2)
sns.violinplot(x="class", y="age", data=titanic)
plt.title("Violin Plot of Age by Class")

# 그래프 표시
plt.show()
```

Q5_0117. 펭귄 데이터셋을 사용하여, 각 종별로 섬(island) 별 개체수를 나타내는 카운트 플롯을 생성하세요. x축은 'island' hue는 'species'로 설정하세요

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 'penguins' 데이터셋 로드
penguins = sns.load_dataset('penguins')

# 카운트 플롯 생성
plt.figure(figsize=(10, 6))
sns.countplot(x='island', hue='species', data=penguins)

# 제목 설정
plt.title('Penguin Count by Island and Species')

# 그래프 표시
plt.show()
```
