```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Q7_0116. 임의의 수치 데이터를 생성하고, 이를 이용해 sin(x) 그래프를 그리세요
- x축과 y축에 적절한 레이블을 추가하세요
- 그래프에 제목을 추가하세요

```python
x = np.arange(0, 10, 0.1)

y = np.sin(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('sin(x) Graph')
plt.show()
```

### Q10_0116. 여러 개의 서로 다른 데이터셋을 생성한 후 이 데이터셋을 사용하여 하나의 창(window)에 여러 개의 서브플롯에 다른 유형의 그래프를 그려보세요(예:선 그래프, 막대그래프, 산점도, 히스토그램)

```python
fig,axes = plt.subplots(2,2,figsize=(10,10))

x = np.arange(160,180,1)
y = np.random.randint(20,80,20)

axes[0,0].plot(np.arange(20),(np.arange(20)*2 + np.random.randn(20)*10), color='black', marker='o', markersize=3)
axes[0,1].hist(np.random.rand(10000), bins=1000, color='red')
axes[1,0].scatter(x,y, color='magenta')
axes[1,1].bar(np.arange(20), np.random.randint(20,80,20))
```
