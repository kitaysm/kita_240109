퍼셉트론
- 인공신경망의 기초가 되는 가장 간단한 형태의 알고리즘 중 하나. 입력(features)에 대해 가중치(weights)를 곱한 후, 이들의 합에 활성화 함수(activation function)를 적용하여 출력(output)을 생성. 퍼셉트론의 목표는 학습 데이터셋을 바탕으로 가중치를 조정하는 것이며, 이를 통해 주어진 입력 데이터를 두 개의 클래스 중 하나로 분류. 1957년에 프랭크 로젠블라트(Frank Rosenblatt)가 제안

- 뉴런은 외부 또는 내부로부터의 자극(예: 빛, 소리, 화학물질, 기계적 압력 등)을 수상돌기에서 수신
- 자극이 충분히 강하면, 뉴런은 액션 포텐셜(action potential)이라는 전기 신호를 생성
- 생성된 액션 포텐셜은 뉴런의 축삭(axon)을 따라 신속하게 전달

- 신경 전달물질은 시냅스를 건너 다음 뉴런의 수상돌기에 위치한 수용체(receptors)와 결합. 이 결합은 다음 뉴런에서 새로운 액션 포텐셜을 유발.

이 과정이 반복되면서 신호는 뇌의 다양한 부분으로 전달



기본 구성 요소
- 입력값(특징)
- 가중치 및 편향: 각 입력 값에 해당 가중치를 곱하고 이러한 곱의 합계에 편향이 추가되어 모델이 보다 유연한 결정을 내리는 데 도움이 된다.
- 활성화 함수: 입력의 가중치 합을 기준으로 뉴런을 활성화할지 여부를 결정하는 함수. 클래식 퍼셉트론의 경우 이는 일반적으로 1 또는 0을 출력하는 계단 함수.

주요 이슈:
1. XOR 문제
  - XOR(배타적 OR) 문제는 단일 레이어 퍼셉트론의 근본적인 결함을 보여준다. 선형적으로 분리 가능한 문제(예: AND 또는 OR 연산)를 처리할 수 있지만 단일 라인으로 입력을 두 클래스로 분리할 수 없는 XOR과 같은 선형으로 분리할 수 없는 문제로 인해 어려움을 겪었다.
  - 이러한 한계는 다층 네트워크의 탐구와 비선형 문제를 해결할 수 있는 다층 퍼셉트론(MLP)의 개발로 이어졌다.
2. 오차 역전파
  - 더 심층적인 네트워크를 훈련하려면 가중치를 업데이트하는 더 효율적인 방법이 필요했다. 1980년대에 도입된 역전파 알고리즘은 다층 네트워크의 효율적인 훈련을 가능하게 했다. 이는 체인 규칙에 따라 각 가중치에 대한 손실 함수의 기울기를 계산하여 네트워크를 통해 출력 계층에서 입력 계층으로 오류를 효과적으로 "전파"하는 방식으로 작동한다.
  이 방법을 사용하면 네트워크 출력의 전체 오류를 최소화하는 방식으로 가중치를 조정할 수 있다.
3. 그라데이션 소멸 문제
  - 네트워크가 깊어짐에 따라 훈련 중에 레이어를 통해 다시 전파됨에 따라 기울기가 점점 더 작아지는 기울기 소멸 문제에 직면. 이로 인해 초기 레이어의 가중치를 업데이트하기가 어려워지고 딥 네트워크 훈련 속도가 느려지거나 심지어 중단되기도 한다.
  - 대체 활성화 함수(예: ReLU) 도입, 정규화 기술 및 특수 네트워크 아키텍처(예: 순환 네트워크용 LSTM)와 같은 혁신이 이 문제를 완화하는 데 도움이 되었다.
  - 활성화 함수의 선택은 네트워크가 복잡한 패턴을 학습하는 능력에 중요한 역할을 한다. 초기 네트워크는 간단한 계단 함수를 사용했지만 입력과 출력 간의 비선형 매핑을 학습할 수 있는 비선형 함수로 대체되었다.
  일반적인 활성화 함수에는 Sigmoid, tanh 및 ReLU(Rectified Linear Unit)가 포함. 각각에는 고유한 장점과 사용 사례가 있으며, ReLU와 그 변형(예: Leaky ReLU, ELU)은 Vanishing Gradient 문제를 완화하는 효율성과 효과로 인해 딥 러닝에서 특히 인기가 있다.

```python
import numpy as np

# 가중치와 바이어스
w11 = np.array([-2,-2])
w12 = np.array([2,2])
w2 = np.array([1,1])
b1 = 3
b2 = -1
b3 = -1

# 퍼셉트론
def MLP(x, w, b):
  y = np.sum(w* x) + b
  if y <=0:
    return 0
  else:
    return 1

# NAND 게이트
def NAND(x1,x2):
  return MLP(np.array([x1, x2]), w11, b1)

# OR 게이트
def OR(x1, x2):
  return MLP(np.array([x1, x2]), w12, b2)

# AND 게이트
def AND(x1, x2):
  return MLP(np.array([x1, x2]), w2, b3)

# XOR 게이트
def XOR(x1, x2):
  return AND(NAND(x1, x2),OR(x1, x2))

# x1, x2 값을 번갈아 대입해 가며 최종 값 출력
for x in [(0,0), (1,0), (0,1), (1,1)]:
  y = XOR(x[0], x[1])
  print("입력 값: "+ str(x) + "출력 값: " + str(y))
```

기울기 업데이트와 미분
- 기계 학습, 특히 경사 하강법(Gradient Descent)과 같은 최적화 알고리즘에서 중요한 역할.
- 이들은 모델을 학습시키고, 손실 함수를 최소화하여 최적의 모델 파라미터를 찾는 데 필수적.

기울기 업데이트의 필요성
- 파라미터 최적화: 기울기 업데이트는 모델의 파라미터(예: 가중치)를 조정하여 손실 함수의 값을 줄이는 과정. 이를 통해 모델의 성능을 최적화할 수 있다.
- 학습 방향 결정: 기울기는 손실 함수의 가장 가파른 상승 방향을 나타내므로, 이의 반대 방향으로 파라미터를 업데이트함으로써 손실을 감소시키고 모델의 예측 정확도를 향상시킬 수 있다.
- 효율적인 학습: 적절한 기울기 계산과 업데이트를 통해, 모델은 불필요한 계산을 줄이고 더 빠르게 최적의 솔루션에 도달할 수 있다.

미분의 필요성
- 손실 함수의 기울기 계산: 미분은 함수의 순간 변화율을 계산. 기계 학습에서 미분을 사용하여 각 파라미터에 대한 손실 함수의 기울기를 구함으로써, 파라미터를 어느 방향으로 조정해야 손실을 줄일 수 있는지 결정할 수 있다.
- 최적화 방향 제공: 미분을 통해 계산된 기울기는 손실을 최소화하기 위해 파라미터를 업데이트하는 데 필요한 방향과 크기를 제공.
- 안정적인 학습 과정: 적절한 미분 계산을 통해 학습률(learning rate)과 같은 하이퍼파라미터를 더 잘 조절할 수 있으며, 이는 학습 과정을 안정적으로 만들고 오버슈팅(overshooting) 또는 느린 수렴과 같은 문제를 방지할 수 있다.

기울기 업데이트와 미분은 모델을 학습시키고 최적화하는 데 있어 상호 보완적인 역할을 한다. 미분을 통해 계산된 기울기를 사용하여 가중치를 업데이트하고, 이 과정을 반복함으로써, 모델은 점차 최적의 파라미터로 수렴하게 된다. 이러한 과정은 다양한 기계 학습 모델과 알고리즘에서 핵심적인 메커니즘으로 작용.

```python
import matplotlib.pyplot as plt
import numpy as np

# 폰트 및 마이너스 기호 설정
plt.rc('font', size=12) # 로컬 환경에 따라 'family' 설정 필요
plt.rcParams['axes.unicode_minus'] = False

# 주어진 함수와 도함수 정의
def f(x):
  """주어진 함수 f(x) = x^2"""
  return x**2

def f_prime(x):
  """함수 f의 도함수 f'(x) = 2x"""
  return 2*x

def tangent_line(f_prime, x, x1, y1):
  """점 (x1, y1)에서 함수 f의 접선의 방정식 y = mx + b에서 x = x-x1, b = y1 (x=>x1일 때 y=>y1)"""
  return f_prime(x1) * (x-x1) + y1

def plot_function_and_tangents(x, x_points):
  """함수와 접선들을 그리는 함수, x_points에서의 접선 계산 및 그리기 로직 포함"""
  fig = plt.figure(figsize=(6,6))
  plt.scatter(x_points, [f(x1) for x1 in x_points], c='k') # 점들 표시
  plt.plot(x, f(x), 'k', label='$f(x) = x^2$') # 함수 f(x) 그래프

  # 각 x_points에서의 접선 계산 및 그리기
  for x1 in x_points:
    y1 = f(x1)
    tangent = tangent_line(f_prime, x, x1, y1) # 접선 계산
    plt.plot(x, tangent, label=f'Tangent at x={x1}') # 접선 그리기

  plt.axhline(0, lw=2, color='k') # x축
  plt.axvline(0, lw=2, color='k') # y축
  plt.ylim(-1,9)
  plt.xlim(-3,3)
  plt.xlabel('x', fontsize=14)
  plt.ylabel('y', fontsize=14)
  plt.grid(lw=1)
  plt.legend()
  plt.show()

# x 좌표 배열과 접선을 그릴 x의 점들 정의
x = np.linspace(-3, 3, 100)
x_points = [1, 2] # 접선을 추가할 x의 점들
plot_function_and_tangents(x, x_points)
```

위의 그래프에는 $x=1$과 $x=2$에서의 접선을 시각화한 것이다.
이 그래프는 기울기 업데이트와 미분의 관계를 잘 보여준다. 여기서 미분은 함수 $f(x)$의 기울기를 $x$의 각 점에서 계산하는 것이며, $x=1$에서의 기울기는 $f'(x) = 2x$로 계산될 때 $2$이고 이 기울기는 접선 $g(x) = 2x - 1$의 기울기와 일치하며, $x=1$에서의 기울기가 바로 이 접선의 기울기이다. 검은색 곡선은 $f(x) = x^2$의 그래프이며, 파란색 선은 $x=1$에서의 접선($g(x) = 2x - 1$), 오랜지색 선은 $x=2$에서의 접선을 나타낸다. 이러한 시각화는 $x=2$에서의 접선의 기울기가 $f'(x) = 2x$로 계산될 때 $4$임을 보여주며, 이는 접선의 기울기가 $x$의 값에 따라 어떻게 변화하는지를 잘 보여준다.

머신러닝에서의 기울기 업데이트는 이러한 미분 개념을 활용하여 모델의 가중치를 조정한다. 모델의 손실 함수의 기울기를 계산하고, 이 기울기의 반대 방향으로 가중치를 조금씩 업데이트하여 손실을 최소화하는 방향으로 모델을 조정하는 것이다.

### 회귀선의 기울기 vs 손실함수가 최소화되는 기울기
- 회귀선의 기울기: 회귀분석에서, 회귀선의 기울기는 독립변수(X)가 종속변수(Y)에 얼마나 큰 영향을 미치는지를 나타내는 값입니다. 예를 들어, 선형 회귀 모델 Y=mx+b에서 m은 회귀선의 기울기로, X의 한 단위 변화가 Y에 미치는 평균적인 영향을 나타낸다. 각 독립 변수에 대한 가중치는 해당 변수가 예측값에 미치는 영향의 크기를 나타낸다.

- 손실 함수가 최소화되는 기울기: 이 기울기는 모델의 파라미터(예: 선형 회귀의 기울기 m과 절편 b)에 대한 손실 함수의 기울기입니다. 이 경우의 '기울기'는 손실 함수의 값을 최소화하기 위해 파라미터를 어떻게 조정해야 하는지를 나타내는 방향과 크기를 가리킨다. 손실 함수는 모델 예측과 실제 데이터 간의 차이를 측정하는 데 사용되며, 기계 학습에서는 이 손실 함수의 기울기(즉, 그래디언트)를 사용하여 모델 파라미터를 업데이트한다. 손실 함수의 기울기는 손실 함수를 각 가중치로 미분한 값입니다. 즉, 각 가중치에 대해 손실 함수가 어떻게 변화하는지를 나타내는 벡터입니다. 이 기울기는 손실을 줄이기 위해 가중치를 어느 방향으로 얼마나 조정해야 하는지를 알려준다.

### 경사하강법(Gradient Descent)
- 머신러닝과 딥러닝에서 비용 함수(Cost Function) 또는 손실 함수(Loss Function)의 값을 최소화하기 위해 사용되는 최적화 알고리즘. 이 방법은 주어진 함수의 최소값을 찾기 위해 반복적으로 파라미터를 조정해 나가는 방식으로 작동.
- 경사하강법의 기본 아이디어는 함수의 기울기(경사)를 계산하고, 그 기울기가 감소하는 방향으로 파라미터를 업데이트하는 것이다. 기울기는 함수의 현재 위치에서 가장 가파르게 증가하는 방향을 나타내므로, 반대 방향으로 이동하면 함수의 최소값을 찾을 수 있다.

```python
# 공부 시간 X와 성적 Y의 넘파이 배열을 만듭니다.
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

# 기울기 a와 절편 b의 값을 초기화합니다.
a = 0
b = 0

# 학습률을 정합니다.
lr = 0.03

# 몇 번 반복될지를 설정합니다.
epochs = 2001

# x 값이 총 몇 개인지 셉니다.
n = len(x)

# 경사 하강법을 시작합니다.
for i in range(epochs):           # epoch 수 만큼 반복

  y_pred = a * x + b              # 예측 값을 구하는 식입니다.
  error = y - y_pred              # 실제 값과 비교한 오차를 error로 놓습니다

  a_diff = (2/n) * sum(-x * (error))  # 오차 함수를 a로 편미분한 값입니다.
  b_diff = (2/n) * sum(-(error))      # 오차 함수를 b로 편미분한 값입니다.

  a = a - lr * a_diff                # 학습률을 곱해 기존의 a 값을 업데이트합니다
  b = b - lr * b_diff                # 학습률을 곱해 기존의 b 값을 업데이트합니다

  if i % 100 == 0:                   # 100번 반복될 때마다 현재의 a 값, b 값을 출력합니다
    print("epoch=%.f, 기울기=%.04f, 절편=%.04f" % (i, a, b))

```

```python
# 앞서 구한 최종 a 값을 기울기, b 값을 y절편에 대입하여 그래프를 그립니다.
y_pred = a * x + b
print(a, b)
# 그래프 출력
plt.scatter(x, y)
plt.plot(x, y_pred, 'r')
plt.show()
```

### 오차 역전파(Backpropagation)
신경망을 훈련시키기 위해 사용되는 핵심 알고리즘 중 하나. 이 방법은 신경망의 출력(output)과 실제 값(target) 사이의 오차를 계산하고, 이 오차를 신경망을 거슬러 올라가며 각 뉴런의 가중치를 조정하는 방식으로 작동.<BR>
오차역전파는 기본적으로 경사 하강법(Gradient Descent)의 원리를 사용하여 신경망의 가중치를 최적화.<BR>

오차 역전파는 두 단계로 이루어진다: 순전파(Forward Pass)와 역전파(Backward Pass).
- 순전파(Forward Pass):
  - 입력 데이터는 신경망의 각 층을 순차적으로 통과하면서, 각 뉴런의 가중치와 활성화 함수를 통해 처리.
  - 마지막 층에서의 출력값은 신경망의 예측값이 된다.
- 역전파(Backward Pass):
  - 신경망의 예측값과 실제 값 사이의 오차를 계산.
  - 이 오차를 바탕으로, 오차 함수의 기울기(미분값)를 계산하고, 이 기울기 정보를 신경망을 거슬러 올라가며 전달.
  - 각 층의 가중치는 계산된 기울기를 사용하여 조정. 이 과정은 신경망의 예측 오차를 줄이는 방향으로 가중치를 업데이트.

중요성
  - 오차역전파는 심층 신경망(Deep Neural Networks)의 훈련에 필수적인 기법으로 이 알고리즘을 통해 복잡한 신경망 구조에서도 효율적으로 가중치를 조정할 수 있다.
  - 이 방법은 신경망의 각 층에서 발생하는 오차의 기여도를 정확히 계산하고, 이를 통해 신경망의 학습이 어떻게 이루어지는지 이해할 수 있게 한다.

오차역전파를 통한 학습 과정에서 적절한 학습률과 초기 가중치 선택이 중요. 잘못된 선택은 학습이 느려지거나, 지역 최소값에 빠지는 등의 문제를 일으킬 수 있다.

신경망의 깊이가 깊어질수록 기울기 소실(Vanishing Gradient) 또는 기울기 폭발(Exploding Gradient) 문제가 발생할 수 있으며, 이를 해결하기 위한 다양한 기법들(예: 가중치 초기화 방법, 정규화 기법, 배치 정규화 등)이 제안되고 있다.

### 기울기 소실(Vanishing Gradient)
- 심층 신경망(Deep Neural Networks)을 훈련시킬 때 발생할 수 있는 현상으로, 신경망의 깊이가 깊어질수록 오차역전파 과정에서 계산된 기울기(오차의 미분값)가 점점 작아져, 입력층에 가까운 층의 가중치가 거의 또는 전혀 업데이트되지 않는 문제를 의미.
- 이 현상은 특히 활성화 함수로 시그모이드(Sigmoid)나 하이퍼볼릭 탄젠트(Hyperbolic Tangent)와 같은 함수를 사용할 때 두드러진다.
- 기울기 소실 문제의 주된 원인은 신경망을 거슬러 올라가며 기울기를 곱해 나가는 과정에서, 기울기 값이 점점 줄어들어 0에 가까워지기 때문입니다. 특히, 활성화 함수의 도함수 값이 0과 1 사이에 있을 때(예: 시그모이드 함수), 신경망이 깊어질수록 기울기가 점점 더 작아지게 된다.
- 기울기 소실 문제가 발생하면, 신경망의 초기 층의 가중치가 충분히 업데이트되지 않아, 모델의 학습이 제대로 이루어지지 않는다. 이는 신경망이 복잡한 패턴이나 특징을 학습하는 데 어려움을 겪게 만들며, 결국 모델의 성능 저하로 이어진다.
- 기울기 소실 문제를 해결하기 위한 몇 가지 방안이 제안되었다:
  - ReLU 활성화 함수 사용: ReLU(Rectified Linear Unit)와 같은 활성화 함수는 양의 입력에 대해 기울기가 1이 되므로, 기울기 소실 문제를 어느 정도 완화할 수 있다.
  - 가중치 초기화 방법 개선: 효과적인 가중치 초기화 방법을 사용하여 신경망의 초기 가중치를 적절히 설정.
  - 정규화 기법 사용: 배치 정규화(Batch Normalization) 등의 정규화 기법을 사용하여 각 층의 입력 분포를 안정화시키고, 기울기 소실 문제를 완화.
  - 잔차 연결(Residual Connections): 잔차 네트워크(ResNet)와 같이 잔차 연결을 통해 입력을 출력에 직접 더함으로써, 신경망이 깊어져도 기울기가 잘 전파될 수 있도록 한다.

### 텐서플로우(TensorFlow)의 케라스(Keras)
- 머신러닝 모델을 쉽고 빠르게 개발할 수 있도록 도와주는 고수준의 신경망 API. 원래 케라스는 독립적인 딥러닝 라이브러리로 시작되었지만, 텐서플로우 1.0 버전부터는 텐서플로우의 공식 프론트엔드로 통합되었고, 텐서플로우 2.0 이후부터는 텐서플로우의 기본 API로 자리잡았다.
- 케라스는 사용자 친화적이며 모듈식으로 구성되어 있어, 신경망의 레이어(layer), 활성화 함수(activation function), 손실 함수(loss function) 등을 마치 레고 블록을 조립하듯 쉽게 조합하여 모델을 구성할 수 있다. 이를 통해 컴퓨터 비전, 자연어 처리와 같은 다양한 머신러닝 애플리케이션을 신속하게 개발할 수 있다.

주요 특징

- 사용자 친화성: 케라스는 일관된 API를 제공하며, 일반적인 사용 사례에 대한 간결한 코드를 지향합니다. 이를 통해 사용자가 직관적으로 이해하고 사용할 수 있다.
- 모듈성: 모델은 독립적인 모듈, 즉 레이어, 손실 함수, 활성화 함수, 최적화 알고리즘 등의 조합으로 구성됩니다. 이러한 모듈은 최소한의 제약으로 서로 연결될 수 있어, 매우 유연한 모델 설계가 가능.
- 확장성: 케라스는 새로운 레이어나 기타 모듈을 만들어 추가하는 것이 상대적으로 쉽습니다. 따라서 연구 목적으로 새로운 아이디어를 실험하고자 할 때 매우 유용.
- 파이썬 기반: 케라스는 순수 파이썬으로 작성되어 있으며, 딥러닝 모델을 파이썬 코드로 쉽게 작성할 수 있다.

기본 구성 요소
- 모델: Sequential 모델과 함수형 API를 통해 신경망을 구성할 수 있다. - Sequential 모델은 레이어를 순차적으로 쌓는 간단한 방식이며, 함수형 API는 더 복잡한 모델을 구성할 수 있는 유연성을 제공.
- 레이어: 다양한 유형의 레이어(예: Dense(완전 연결 레이어), Conv2D(2D 컨볼루션 레이어), RNN(재귀 신경망 레이어))를 제공.
- 손실 함수: 모델의 예측과 실제 값 사이의 차이를 측정하는 함수입니다. 예를 들어, 회귀 문제에는 mean_squared_error, 분류 문제에는 categorical_crossentropy가 자주 사용.
- 옵티마이저: 경사하강법 알고리즘을 구현한 것으로, adam, sgd, rmsprop 등이 있다. 모델의 손실을 최소화하기 위해 모델 파라미터를 업데이트하는 방법을 정의.
- 평가 지표: 학습과 테스트 단계에서 모델의 성능을 평가하기 위해 사용되는 지표들이다. 예를 들어, 정확도(accuracy), 정밀도(precision), 재현율(recall) 등이 있다.

케라스는 이러한 구성 요소들을 사용하여 다양한 신경망 아키텍처를 구현하고, 이를 텐서플로우 백엔드를 통해 효율적으로 실행할 수 있게 해준다.

### 텐서플로우(TensorFlow)에서 모델을 구축 방법
Sequential API와 함수형(Functional) API. 이 두 방법은 신경망 모델을 정의, 구축 및 실험하는 데 사용되며, 사용자의 필요와 모델의 복잡성에 따라 적합한 방법을 선택할 수 있다.

Sequential API
- Sequential API는 가장 간단한 형태의 모델을 생성하는 방법 중 하나로, 레이어를 순차적으로 쌓아 올리는 방식으로 모델을 정의. 각 레이어는 바로 이전 레이어의 출력을 입력으로 받아 처리.
- Sequential 모델은 단순한 순차적 구조를 가진 신경망에 적합하며, 복잡한 모델 구조(예: 다중 입력/출력, 모델 내 레이어 간 병렬 연결 등)를 표현하는 데는 한계가 있다.

함수형(Functional) API
- 함수형 API는 더 유연한 모델 설계를 가능하게 하는 API로, 복잡한 모델 아키텍처를 구성할 수 있다.
- 함수형 API를 사용하면 다중 입력 및 출력, 공유 레이어(동일한 레이어를 여러 번 사용), 레이어 간 복잡한 연결 구조 등을 정의할 수 있다. 이는 각 레이어의 입출력을 명시적으로 정의함으로써 달성된다. 함수형 API는 고급 사용자에게 더 많은 유연성과 제어력을 제공한다.

[ 활성화 함수 ]
- ReLU(Rectified Linear Unit): 'f(x) = max(0, x)'로 정의되어 널리 사용되는 활성화 함수입니다. 계산 효율성이 높으면서도 비선형성이 도입되어 훈련 중에 모델이 빠르게 수렴될 수 있다.
- 시그모이드: 이 활성화 함수는 0과 1 사이의 값을 출력하므로 이진 분류 문제에 적합.
- 소프트맥스: 각 클래스에 대한 확률을 얻기 위해 분류기의 출력 레이어에서 자주 사용되는 소프트맥스 함수는 여러 클래스에 대한 확률 분포를 출력.
- Tanh(하이퍼볼릭 탄젠트): 시그모이드와 비슷하지만 -1과 1 사이의 값을 출력.

- 최적화의 목표는 비용 함수(또는 손실 함수)를 최소화하는 모델에 대한 최상의 매개변수(가중치)를 찾는 것이다. 이 프로세스는 모델이 특정 작업에 대한 정확성과 성능을 향상시키는 데 필수적이다.

[ 옵티마이저 유형 ]

가중치를 조정하기 위한 자체 메커니즘과 전략을 가진 여러 가지 최적화 프로그램이 있습니다. 가장 일반적으로 사용되는 최적화 도구는 다음과 같다.

- 확률적 경사하강법(SGD): 이것은 가장 간단한 최적화 프로그램 중 하나. 각 훈련 예제를 개별적으로 사용하여 모델의 가중치를 업데이트. 여기에는 각 단계에서 훈련 데이터의 하위 집합을 사용하여 더 효율적으로 만드는 미니 배치 경사하강법(Mini-Batch Gradient Descent)과 같은 변형이 있다.
- 모멘텀: 이 최적화 프로그램은 관련 방향을 따라 탐색하여 SGD를 가속화하고 현재 업데이트 벡터에 과거 단계의 업데이트 벡터의 일부를 추가하여 높은 곡률 방향의 진동을 완화.
- Adagrad: 학습 속도를 매개변수에 맞게 조정하여 자주 발생하는 기능과 관련된 매개변수에 대해 더 작은 업데이트를 수행하고, 자주 발생하지 않는 기능과 관련된 매개변수에 대해 더 큰 업데이트를 수행.
- RMSprop: 이 최적화 프로그램은 가중치에 대한 학습률을 해당 가중치에 대한 최근 기울기 크기의 이동 평균으로 나눈다. 이는 Geoff Hinton이 강의에서 제안한 미공개 적응형 학습률 방법.
- Adam(Adaptive Moment Estimation): AdaGrad 및 RMSprop 알고리즘의 최상의 속성을 결합하여 노이즈 문제에 대한 희소 기울기를 처리할 수 있는 최적화 알고리즘을 제공.

[ 옵티마이저 작동 방식 ]

- 최적화 프로그램은 모델 매개변수에 대한 손실 함수의 기울기를 사용하여 조정. 이 기울기는 손실을 최소화하기 위해 가중치를 조정해야 하는 방향을 나타낸다. 가중치를 반복적으로 업데이트함으로써 최적화 프로그램은 모델의 최고 성능에 해당하는 손실 함수의 가장 낮은 지점에 도달하려고 한다.
- 최적화 프로그램의 선택은 신경망의 성능에 큰 영향을 미칠 수 있으며, 최선의 선택은 특정 문제, 데이터의 성격, 신경망 아키텍처에 따라 달라질 수 있다. Adam은 다양한 문제에 대한 적응성과 성능으로 인해 기본적으로 좋은 선택이 되는 경우가 많지만, 특정 작업과 데이터 세트에는 다른 최적화 프로그램이 도움이 될 수 있다. 주어진 시나리오에 가장 효과적인 최적화 프로그램을 식별하려면 일반적으로 실험과 교차 검증이 필요하다.

[ 손실 함수라고도 알려진 비용 함수 ]
- 모델의 예측 출력과 실제 목표 값 간의 차이를 측정하는 수학적 함수입니다. 이는 모델의 오류를 정량화하여 훈련 중 최적화 프로세스에 대한 가이드 역할을 한다.
- 신경망 훈련의 목표는 이 비용 함수를 최소화하는 것이다. 이는 모델이 예측하는 것과 실제로 관찰되는 것 사이의 오류를 효과적으로 줄이는 것을 의미.
- 고급 신경망 API인 Keras는 다양한 유형의 문제에 대해 다양한 비용 함수를 내장하고 있으며 필요한 경우 사용자 정의 손실 함수를 생성할 수도 있다. 비용 함수의 선택은 분류, 회귀 등과 같이 해결되는 문제의 구체적인 성격에 따라 달라다.

[ Keras의 일반적인 비용 함수 ]
- 평균 제곱 오차(MSE): 회귀 문제에 사용됩니다. 예측값과 실제값 사이의 오차 제곱의 평균을 계산. keras.losses.MeanSquaredError()
- 이진 교차엔트로피: 이진 분류 문제에 사용됩니다. 두 확률 분포, 즉 실제 레이블 분포와 예측 확률 사이의 거리를 측정. keras.losses.BinaryCrossentropy()
- 범주형 교차엔트로피: 레이블이 원-핫 인코딩되는 다중 클래스 분류 문제에 사용. 이진 교차엔트로피와 유사하지만 여러 클래스에 적용. keras.losses.keras.losses.CategoricalCrossentropy()
- 희소 범주형 교차엔트로피: 범주형 교차엔트로피와 유사하지만 레이블이 정수인 경우에 사용됩니다. 클래스가 많은 경우에는 메모리 효율성이 더 높다. keras.losses.SparseCategoricalCrossentropy()

[ 배치 크기 ]
- 배치 크기는 기계 학습에 사용되는 용어로, 모델 학습 프로세스의 한 번의 반복에서 활용되는 학습 예제의 수를 나타낸다.
- 신경망 훈련의 맥락에서 데이터 세트는 일반적으로 여러 배치로 나뉘며 모델의 가중치는 각 배치를 처리한 후 업데이트된다.
- 배치 크기의 선택은 수렴 속도와 훈련 프로세스의 안정성을 포함한 훈련 역학에 큰 영향을 미칠 수 있다.

- 작은 배치 크기: 각 업데이트 내에서 더 많은 노이즈로 더 빠른 업데이트가 가능. 이는 때때로 모델이 로컬 최소값을 벗어나는 데 도움이 될 수 있지만 훈련 과정을 더욱 불안정하게 만들 수도 있다.
- 대형 배치 크기: 기울기에 대한 더 정확한 추정치를 제공하지만 업데이트 시 노이즈가 적다. 이는 더 안정적이지만 잠재적으로 더 느린 훈련으로 이어질 수 있습니다. 또한 대규모 배치에는 더 많은 메모리가 필요하므로 일부 하드웨어에서는 제한 요소가 될 수 있다.

[ 에포크 ]
- 에포크는 전체 훈련 데이터세트를 한 번 통과하는 것을 나타낸다. 한 epoch 동안 신경망은 모든 훈련 예제를 한 번 처리한다. 따라서 훈련의 에포크 수에 따라 학습 알고리즘이 전체 훈련 데이터 세트에서 작동하는 횟수가 결정된다.
- 너무 적은 에포크를 실행하면 모델이 훈련 데이터에서 충분히 학습하지 못하는 과소적합이 발생할 수 있다.
- 너무 많은 에포크를 실행하면 과적합이 발생할 수 있다. 즉, 모델이 노이즈를 포함하여 훈련 데이터에서 패턴을 너무 잘 학습하여 보이지 않는 데이터에 대한 성능이 저하될 수 있다.

[ 배치 크기와 에포크 간의 관계 ]
- 배치 크기 영향: 배치 크기는 가중치를 업데이트하기 전에 확인하는 예시 수를 결정. 배치가 작을수록 한 번에 더 많은 업데이트가 이루어지며 잠재적으로 학습 속도가 빨라지지만 너무 작으면 불안정해질 수 있다. 배치가 클수록 더 안정적인 기울기 추정이 제공되지만 학습 속도가 느려지고 더 많은 메모리가 필요할 수 있다.
- 에포크 영향: 에포크 수에 따라 학습 알고리즘이 전체 데이터 세트를 순환하는 횟수가 결정. Epoch가 많을수록 모델이 가중치를 학습하고 조정할 수 있는 기회가 더 많아지지만 숫자가 너무 높으면 과적합이 발생할 위험이 있다.
- 요약하자면, 배치 크기와 에포크 수는 신중하게 균형을 맞춰야 하는 신경망 훈련의 두 가지 기본 하이퍼파라미터이다. 이러한 매개변수의 최적 설정은 특정 데이터세트와 문제는 물론 사용 가능한 계산 리소스에 따라 달라진다.

[ 완전신경망(완전연결신경망, Fully Connected Neural Network)에서 각 층의 파라미터 수 계산 ]

- 각 연결에 하나의 가중치가 있으므로, 특정 층의 가중치 수는 이전 층의 뉴런 수와 해당 층의 뉴런 수를 곱한 값과 같다.
- 각 뉴런에는 하나의 편향이 있으므로, 특정 층의 편향 수는 해당 층의 뉴런 수와 같다.
- 특정 층의 총 파라미터 수는 가중치 수와 편향 수의 합이다.

가중치의 의미
- 이전 층의 뉴런 수: 각 뉴런은 이전 층의 모든 뉴런으로부터 입력 신호를 받는다. 이전 층에 더 많은 뉴런이 있을수록, 현재 층의 한 뉴런이 처리해야 할 입력 신호의 수가 더 많아진다.
- 해당 층의 뉴런 수: 해당 층에 더 많은 뉴런이 있다는 것은, 이전 층의 각 뉴런으로부터 오는 신호를 받아들이는 더 많은 처리 단위가 있다는 것을 의미.
- 가중치의 역할: 각 가중치는 입력 신호의 중요도를 조정. 즉, 특정 입력이 해당 뉴런의 활성화에 얼마나 영향을 미치는지 결정하는 역할을 한다. 가중치는 학습 과정에서 조정되며, 신경망이 훈련 데이터로부터 패턴을 학습하는 방식을 결정.

가중치 수 계산
- 신경망에서 정보가 전달되는 방식과 학습이 어떻게 이루어지는지를 나타낸다. 각 가중치는 이전 층의 특정 뉴런과 현재 층의 특정 뉴런 사이의 연결 강도를 나타내며 이전 층의 뉴런 수와 해당 층의 뉴런 수를 곱하는 것은 모든 가능한 연결을 고려하여 이들 간의 관계를 정량화하는 방법이다.
- 계산 과정을 통해 각 연결마다 하나의 고유한 가중치가 할당




이러한 구조는 신경망이 복잡한 패턴과 관계를 모델링할 수 있도록 하며, 다양한 입력에 대한 반응을 조정하는 데 필요한 유연성을 제공합니다. 가중치의 조정은 신경망이 학습 과정에서 경험을 통해 얻은 지식을 내재화하는 방식입니다.

[ 픽셀(pixel) ]
- 디지털 이미지를 구성하는 기본 단위. 단어 "픽셀"은 "Picture Element"의 줄임말로, 이미지를 구성하는 가장 작은 단일 색상의 점을 의미. 각 픽셀은 특정 색상과 밝기 값을 가지며, 이러한 픽셀들이 모여 전체 이미지를 형성
- 색상과 밝기: 컬러 이미지에서 각 픽셀은 일반적으로 빨강(Red), 초록(Green), 파랑(Blue)의 세 가지 색상 채널 값을 가지며, 이를 RGB 값이라고 한다. 각 채널의 값은 보통 0에서 255 사이의 정수로 표현되며, 이 값들의 조합을 통해 다양한 색상을 나타낸다. 그레이스케일 이미지에서는 단일 채널만 사용하여 밝기를 표현하며, 0은 검은색, 255는 흰색을 의미하고, 그 사이의 값은 다양한 회색 음영을 나타낸다.
- 해상도: 이미지의 해상도는 이미지의 세부 사항과 선명도를 결정하는 중요한 요소입니다. 해상도가 높은 이미지는 더 많은 픽셀을 가지며, 따라서 더 세밀한 디테일을 표현할 수 있다. 예를 들어, 1920x1080 해상도의 이미지는 가로 1920픽셀, 세로 1080픽셀로 구성되어 있다.
- 디지털 표현: 디지털 이미지 처리, 컴퓨터 그래픽, 디지털 카메라 촬영과 같은 분야에서 픽셀은 기본적인 작업 단위로 사용되며 이미지의 품질, 색상 처리, 이미지 분석 등은 모두 픽셀 데이터를 기반으로 한다.
