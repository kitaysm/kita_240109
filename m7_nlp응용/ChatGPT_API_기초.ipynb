### OpenAI의 텍스트 생성 모델
- 종종 사전 훈련된 생성 변환기 또는 대규모 언어 모델이라고도 하며 자연어, 코드 및 이미지를 이해하도록 훈련되었습니다.
- 모델은 입력에 대한 응답으로 텍스트 출력을 제공합니다. 이러한 모델에 대한 입력을 "프롬프트"라고도 합니다. 프롬프트 디자인은 본질적으로 대규모 언어 모델 모델을 "프로그래밍"하는 방법으로, 일반적으로 작업을 성공적으로 완료하는 방법에 대한 지침이나 몇 가지 예를 제공.
- OpenAI API를 통해 이러한 모델 중 하나를 사용하려면 입력과 API 키가 포함된 요청을 보내고 모델의 출력이 포함된 응답을 받게 됩니다. 최신 모델 gpt-4및 gpt-3.5-turbo는 채팅 완료 API 엔드포인트를 통해 액세스됩니다.

### Chat Completions API
- 채팅 모델은 메시지 목록을 입력으로 사용하고 모델 생성 메시지를 출력으로 반환합니다. 채팅 형식은 여러 차례 대화를 쉽게 할 수 있도록 설계되었지만 대화가 없는 단일 작업에도 마찬가지로 유용합니다.
- 주요 입력은 message 매개변수입니다. 메시지는 메시지 개체의 배열이어야 하며, 각 개체에는 역할( "system", "user", "assistant")과 콘텐츠가 있습니다. 대화는 하나의 메시지만큼 짧을 수도 있고 여러 번 주고받을 수도 있습니다.
- 일반적으로 대화는 먼저 시스템 메시지로 형식화되고 이어서 사용자 메시지와 보조자 메시지가 교대로 표시됩니다.
- 시스템 메시지는 어시스턴트의 동작을 설정하는 데 도움이 됩니다. 예를 들어, 어시스턴트의 성격을 수정하거나 대화 전반에 걸쳐 어시스턴트가 어떻게 행동해야 하는지에 대한 구체적인 지침을 제공할 수 있습니다. 그러나 시스템 메시지는 선택 사항이며 시스템 메시지가 없는 모델의 동작은 "You are a helpful assistant."와 같은 일반적인 메시지를 사용하는 것과 유사할 가능성이 높습니다.
- 사용자 메시지는 보조자가 응답할 요청이나 설명을 제공합니다. 보조 메시지는 이전 보조 응답을 저장하지만 원하는 동작의 예를 제공하기 위해 사용자가 작성할 수도 있습니다.

### get_completion 사용자 함수

````python
def get_completion(prompt, model="gpt-3.5-turbo"):
  messages = [{"role": "user", "content": prompt}]
  response = client.chat.completions.create(
      model=model,
      messages=messages,
      temperature=0.1,
      max_tokens=100,
      top_p=1.0,
      )
  return response.choices[0].message.content
````

### Prompting Principles
- Principle 1: Write clear and specific instructions
- Principle 2: Give the model time to “think”
#### Tactics
Tactic 1: Use delimiters to clearly indicate distinct parts of the input
- Delimiters can be anything like: ```, """, < >, <tag> </tag>, :

````python
text = f"""
You should express what you want a model to do by \
providing instructions that are as clear and \
specific as you can possibly make them. \
This will guide the model towards the desired output, \
and reduce the chances of receiving irrelevant \
or incorrect responses. Don't confuse writing a \
clear prompt with writing a short prompt. \
In many cases, longer prompts provide more clarity \
and context for the model, which can lead to \
more detailed and relevant outputs.
"""
prompt = f"""
Summarize the text delimited by triple backticks \
into a single sentence.
```{text}```
"""
response = get_completion(prompt)
print(response)
````

### generate_chat_respons 사용자 함수
